##Slicing: array[start:end:stride,column]

extract 1+ element from an array

the colon : indicates slicing when it is part of an index []

for slicing a 1D array:
```
array[start:end:stride]
```

To slice by row and column, combine the rule of array[0,1] and the above:

```
array[(start:end:stride),2]
```
This returns slicing only in the 3rd column ^




```
a[start],a[start+stride],a[start+2*stride],...,a[start+M*stride]
```

where the last one is < end



````
A = np.zeros((5,9))
A[0]=0
A[1]=1
A[2]=2
A[3]=3
A[4]=4
A[1,1]=8

N=np.size(A,0) # just returns number of rows in A
columnOne=A[0:N:1,0] # I thought this would return a vertical column. does slicing turn arrays into 1D?
columnTwo=A[0:N:1,1]
rowOne=A[0,:]
rowTwoEven=A[1,::2] # in row 1, slice every other number 

print(A)
print(N)
print(columnOne)
print(columnTwo)
print(rowOne)
print(rowTwoEven)
````

## array.ravel() or array.flatten()

```

vs.

```

## np.reshape(array,(3D,r,c)) or array.reshape((3D,r,c))

used to reshape array into any shape consistent with the number of elements

the product of the numbers in the tuple array (3D,r,c) must be equal to the number of elements in the array

DOES NOT CREATE A NEW INDEPENDENT ARRAY. INSTEAD IT STILL HAS ACCESS/INFLUENCE OVER ORIGINAL ARRAY; THUS, IF CHANGE ANY ELEMENTS IN THE RESHAPED ARRAY, WILL ALSO CHANGE IN THE ORIGINAL ARRAY

````
a = np.arange(12) # 0,1,2,3,4,5,6,7...11
b = np.reshape(a,(3,4)) # 3r x 4c
c = b.reshape((2,6)) # 2r x 6c
d = c.reshape((2,3,2)) # 2 array of 3r x 2c
print(a)
print(b)
print(c)
print(d)
````

````
# measurement.py
def distance(pointA,pointB=(0,0),metric='taxi'):
  """
  Return distance in city blocks between points A and B.
  If metric is 'taxi' (or omitted), use taxicab metric/
  Otherwise, use Euclidean distance.
  
  pointA=(x1,y1)
  pointB=(x2,y2)
  
  If point B is omitted, use the origin.
  """
  if metric=='taxi':
    interval=abs(pointB[0]-pointA[0])+abs(pointB[1]-pointA[1])
  else:
    interval=np.sqrt(((pointB[0]-pointA[0])**2)+(pointB[1]-pointA[1])**2)
  return interval
````

## R2 Project! 

Here I created a function that will automatically calculate the R$^2$ value and print out a graph. You just need to input a number from 0–3 to plug in the explanatory variable x. 0=high school GPA, 1=math SAT, 2=verbal SAT, 3=computer science GPA. 

I'm especially proud that the x-label will automatically output the explanatory variable. Also, I'm proud of the x axis scaling because it is different for GPA and SAT so I had to change that to make it based on the explanatory variable.

````
import urllib
import numpy as np
import matplotlib.pyplot as plt
web_file=urllib.request.urlopen('http://onlinestatbook.com/mobile/case_studies/data/sat.txt')
header=np.loadtxt(web_file,encoding='utf-8',delimiter=' ',dtype=str,max_rows =1)
data=np.loadtxt(web_file,encoding='utf-8',delimiter=' ',skiprows=1)
def predict(N):
  """
  Here I created a function that will automatically calculate 
  the R$^2$  value and print out a graph. 

  You just need to input a number from 0–3 to 
  plug in the explanatory variable x. 

  0=high school GPA
  1=math SAT
  2=verbal SAT
  3=computer science GPA
  """
  plt.plot(data[:,N],data[:,4],'+')
  ax=plt.gca()
  ax.set_xlabel(str(header[N]))
  ax.set_ylabel('University GPA')
  ax.set_title(str(header[N])+' as a predictor of University GPA')
  output = np.polyfit(data[:,N],data[:,4],1)
  if N==0 or N==3:
    regline=output[1]+output[0]*np.arange(2,4,.1)
    plt.plot(np.arange(2.0,4,.1),regline)
  else:
    regline=output[1]+output[0]*np.arange(450,800,.1)
    plt.plot(np.arange(450,800,.1),regline)
  y=data[:,4]
  sstot=np.sum((y-np.mean(y))**2)
  x=data[:,N]
  ssres=np.sum((y-(output[1]+output[0]*x))**2)
  r2=1-(ssres/sstot)
  r2print="R² is {:f}".format(r2)
  return(r2print)
````

````
# find the mean-square displacement of a 4000-step walk

# VECTORIZED
def hi(num_steps):
  x_steps=rng(num_steps)
  y_steps=rng(num_steps)
  x_steps=2*x_steps-1
  y_steps=2*y_steps-1
  cumsumx=np.cumsum(x_steps)
  cumsumy=np.cumsum(y_steps)
  displacement=np.sqrt(cumsumx[-1]**2+cumsumy[-1]**2) # pythagorean theorem
  returned =np.mean(displacement**2)
  return returned
hi(4000)
````

````
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.pyplot as plt
import colorsys
from PIL import Image
import pandas as pd
from urllib.request import urlopen

# (1) Import the file to be analyzed!
img_file = Image.open(urlopen("http://hyperphysics.phy-astr.gsu.edu/hbase/Astro/astpic/crab.jpg"))
img = img_file.load()   # this is called a PIXEL ACCESS OBJECT

# (2) Construct a blank matrix representing the pixels in the image
[xs, ys] = img_file.size
max_intensity = 100
hues = {}

# (3) Examine each pixel in the image file and add it to an array
r_ = []
g_ = []
b_ = []
colours = []

for x in range(0, xs):
  for y in range(0, ys):
    # ( )  Get the RGB color of the pixel
    [r, g, b] = img[x, y] # HOW DOES THREE COMPONENTS TURN INTO TWO? 

    # ( )  Normalize pixel color values
    r /= 255.0
    g /= 255.0
    b /= 255.0

    r_.append(r)
    g_.append(g)
    b_.append(b)
    colours.append([r, g, b])
    
    img[x,y] = (int(r*255), 0, 0)   # note by messing with the pixel access object, I'm changing the image!
    
# (4)   Plot the array!
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.scatter(r_, g_, b_, s=5, c=colours, lw=0)

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
# fig.add_axes(ax)
plt.show()
````
